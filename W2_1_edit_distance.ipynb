# Edit Distance

## Exercise 1:

compute the minimum edit distance between:
1. example
1. sample

scenarios:
 - only insertion and substitutions are allowed, costing 1 each
 - also substitution is allowed, costing 1.5

```python
def edit_distance(s1, s2, insertion_cost=1, deletion_cost=1, substitution_cost=1.5):
    n = len(s1) + 1
    m = len(s2) + 1

    # Create a matrix to store the edit distances and operations
    dp = [[(0, None)] * m for _ in range(n)]
    
    # Initialize the first row and column
    for i in range(1, n):
        dp[i][0] = (i * deletion_cost, 'D')  # Deletion
    for j in range(1, m):
        dp[0][j] = (j * insertion_cost, 'I')  # Insertion

    # Fill the matrix
    for i in range(1, n):
        for j in range(1, m):
            if s1[i - 1] == s2[j - 1]:
                cost = 0
                op = 'S'  # Match
            else:
                cost = substitution_cost
                op = 'S'  # Substitution

            candidates = [
                (dp[i - 1][j][0] + deletion_cost, 'L'),      # Deletion
                (dp[i][j - 1][0] + insertion_cost, 'R'),     # Insertion
                (dp[i - 1][j - 1][0] + cost, op)             # Match or Substitution
            ]
            
            dp[i][j] = min(candidates, key=lambda x: x[0])

    # Reconstruct the alignment
    aligned_s1 = []
    aligned_s2 = []
    i, j = n - 1, m - 1
    while i > 0 or j > 0:
        cost, op = dp[i][j]
        if op == 'L':
            aligned_s1.append(s1[i - 1])
            aligned_s2.append('-')
            i -= 1
        elif op == 'R':
            aligned_s1.append('-')
            aligned_s2.append(s2[j - 1])
            j -= 1
        else:  # 'D'
            aligned_s1.append(s1[i - 1])
            aligned_s2.append(s2[j - 1])
            i -= 1
            j -= 1

    # Reverse the alignments
    aligned_s1.reverse()
    aligned_s2.reverse()

    return dp[-1][-1][0], ''.join(aligned_s1), ''.join(aligned_s2)

s1 = "example"
s2 = "sample"

distance, aligned1, aligned2 = edit_distance(s1, s2)
print(f"The edit distance between '{s1}' and '{s2}' is: {distance}")
print(f"Alignment:")
print(f"s1: {aligned1}")
print(f"s2: {aligned2}")
```

## Computing edit distance using nltk

```python
import nltk

# Define 2 strings to compare
string1 = "example"
string2 = "sample"

# Compute the edit distance using the `edit_distance` method in the NLTK library
edit_distance = nltk.edit_distance(string1, string2, substitution_cost=1.5, transpositions=False)

# Print the result
print("The edit distance of the two strings is:", edit_distance)
```

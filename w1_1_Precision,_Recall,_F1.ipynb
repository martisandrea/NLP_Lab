## AnimalFinder
Let's suppose we know there are exactly 123 different types of animals in the world.
We developed _AnimalFinder_, a ML algorithm that read a text and find animals mentioned.
After running it against all nature books (we assume all animals are mentioned), it outputs:

    ['Cat','Rome','Dog','Bird','Tower']

Compute:
1. Precision, Recall, F1-measure of _AnimalFinder_
2. what are FP and TP?
3. how many FN, TN? Provide 2 examples of each

### Answeres:
1. Precision: 3/3+2 -> 3/5 <br>
   Recall: suppose that the total number of ground truth True labels is `x`, then the number of False Negatives is `FN = x-TP = x-3`, then  the recall value is `3/(3+x-TP) = 3/(3+x-3) = 3/x` <br>
   F1-measure: `(2 * 3/5 * 3/x)/(3/5 + 3/x) = (18/5x)/(3/5 + 3/x)` <br>
2. False Positives are words that are not animals but were classified as an animal word, such as 'Rome' in the output list (the outputs should be all animals); True Positives are animal words correctly classified as that, such as 'Cat'.
3. The number of FN, as already said in point 1, is `x-3`; following the same logic, suppose that the the total number of ground truth False labels is `y`, then the number of True Negatives is `TN = y-FP = y-2`. An example of FN is a word like `Dog` non present in the output of the model, while an example of TN is a word like `Tower` not present in the output. 

## Regions in Italy

we have two AI algorithms that can answer a set, given an appropriate question. The question is "List all Regions in Italy". Here are the outputs:

1. AlgA: `['Sardinia','Tuscany']`
2. AlgB: `['Sardinia','Tuscany','Sicily','Rome']`
2. AlgC: `['Sardinia','Tuscany','Sicily','Veneto']`

Which one is better?


### Answeres
Precision of AlgA: `2/2 = 1`
<br>
Recall of AlgA: `2/20 = 0.10`
<br>
<br>
Precision of AlgB: `3/4 =  0.75`
<br>
Recall of AlgB: `3/20 = 0.15`
<br>
<br>
Analizing only precision and recall cannot give as an absolute winner between the algorithms; we could prefer AlgA if we want only correct outputs, AlgB if we want as many correct answeres as possible, having possible wrong elements in the output. Lets compute the F1-measure to check which one is the best.
<br>
<br>
F1-measure of AlgA: `(2 * 1 * 0.10)/(1 + 0.10) = 0.18`
<br>
F1-measure of AlgB: `(2 * 0.75 * 0.15)/(0.75 + 0.15) = 0.25`
<br>
<br>
So AlgB is better than AlgA.

# Ranges and interdependencies of performances

Assume the precision of a model is 80%.
 - What is the maximum and minimum recall
 - What is the maximum and minimum f1-measure?


### Answeres
* Since the Recall metrics is computeng using FN instead of FP (and we can not deduct the number of FN by only knowing the precision value), the maximum value of recall can be `100%` if the number of FN is 0, while the minimum value is almost `0%` if the number of FN is very high.
* F1-measure is an harmonic mean of Precision and Recall, we can compute its maximum value: the maximum is given if recall is 1, so `F1 = (2 * 0.8 * 1)/(0.8 + 1) = 89%`. The minimum value is given if Recall is 0, so `F1 = (2 * 0.8 * 0)/(0.8 + 0) = 0/0.8 = 0%`.
